```python
import tf_keras as keras
import tensorflow_hub as hub
```

```python
model = keras.Sequential([
    keras.layers.Input(shape=(224, 224, 3)),
    keras.layers.Rescaling(1.0 / 255.0), # 0~1
    hub.KerasLayer("https://www.kaggle.com/models/tensorflow/efficientnet/TensorFlow2/b0-classification/1")
])
```

```python
import numpy as np
from PIL import Image
from keras.applications import efficientnet

pistachio = Image.open("Pistachio_Image_Dataset/Kirmizi_Pistachio/kirmizi (1).jpg")
pistachio = pistachio.resize((224, 224))
```

```python
display(pistachio)
```

```python
pistachio = np.array(pistachio)
pistachio = pistachio[np.newaxis, :]
pistachio.shape
```

```python
predictions = model.predict(pistachio)
efficientnet.decode_predictions(predictions)
```
