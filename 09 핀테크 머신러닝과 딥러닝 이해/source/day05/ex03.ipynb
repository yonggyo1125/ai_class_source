```python
import pandas as pd
df = pd.read_csv('https://bit.ly/fish_csv_data')
df.head()
```

```python
import numpy as np

fish_data = df[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
fish_target = df['Species'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(
    fish_data, fish_target, test_size=0.2, random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)

train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(train_scaled, train_target)
```

```python
lr.predict(test_scaled[:5])
```

```python
lr.classes_
```

```python
# 선형 방정식의 해를 구하면...
decision = lr.decision_function(test_scaled[:5])
decision
```

```python
# 분류하려면 각 클래스마다 구해진 해의 비율 - 소프트맥스 
from scipy.special import softmax
np.round(softmax(decision, axis=1), decimals=3)
```

```python
np.round(lr.predict_proba(test_scaled[:5]), decimals=3)
```
