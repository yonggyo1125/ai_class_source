```python
fig, axs = plt.subplots(1, 3)
axs[0].imshow(fruits[0], cmap='gray_r')
axs[1].imshow(fruits[100], cmap='gray_r')
axs[2].imshow(fruits[200], cmap='gray_r')
plt.show()
```

```python
# 사과, 파인애플, 바나나 각각의 픽셀 평균
apple = fruits[0:100].reshape(-1, 100 * 100) # 너비 X 높이 -> 1차원 배열 10000
pineapple = fruits[100:200].reshape(-1, 100 * 100)
banana = fruits[200:300].reshape(-1, 100 * 100)
```

```python
apple.shape
```

```python
np.mean(apple, axis=1) # axis: 0 - 행, 1 - 열
```

```python
np.mean(pineapple, axis=1)
```

```python
np.mean(banana, axis=1)
```

```python
#_, axs = plt.subplots(1, 3, figsize=(20, 5))
plt.figure(figsize=(10,5))
plt.hist(np.mean(apple, axis=1), alpha=0.8)
plt.hist(np.mean(pineapple, axis=1), alpha=0.8)
plt.hist(np.mean(banana, axis=1), alpha=0.8)
plt.legend(['apple', 'pineapple', 'banana'])
plt.show()
```

```python
# 100개 사과 평균 픽셀, 100개의 파인애플 평균 픽셀, 100개의 바나나 평균 픽셀
apple_mean = np.mean(apple, axis=0).reshape(100, 100)
pineapple_mean = np.mean(pineapple, axis=0).reshape(100, 100)
banana_mean = np.mean(banana, axis=0).reshape(100, 100)

_, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(apple_mean, cmap='gray_r')
axs[1].imshow(pineapple_mean, cmap='gray_r')
axs[2].imshow(banana_mean, cmap='gray_r')
plt.show()
```

```python
apple_diff = np.abs(fruits - apple_mean)
apple_diff_mean = np.mean(apple_diff, axis=(1, 2))
apple_diff_mean
```

```python
apple_index = np.argsort(apple_diff_mean)[:100] # 모든 사과의 평균과 차이가 가장 작은 순서대로 정렬하고 그 인덱스 번호 반환 
apple_index
```

```python
_, axs = plt.subplots(10, 10, figsize=(10, 10))
for i in range(10): # 10행
    for j in range(10): # 10열
        # i * 10 + j
        axs[i, j].imshow(fruits[apple_index[i * 10 + j]], cmap='gray_r')
        axs[i,j].axis('off')

plt.show()
```

```python
pineapple_diff = np.abs(fruits - pineapple_mean)
pineapple_diff_mean = np.mean(pineapple_diff, axis=(1,2))
```

```python
pineapple_index = np.argsort(pineapple_diff_mean)[:100]
_, axs = plt.subplots(10, 10, figsize=(10, 10))
for i in range(10): # 10행
    for j in range(10): # 10열
        # i * 10 + j
        axs[i, j].imshow(fruits[pineapple_index[i * 10 + j]], cmap='gray_r')
        axs[i,j].axis('off')

plt.show()
```
