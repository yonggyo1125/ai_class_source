```python
import requests
html = requests.get("https://finance.naver.com/marketindex/exchangeList.naver").text
html
```

```python
data = html.split("미국 USD")[1].strip()
data = data.split("유럽연합 EUR")[0].strip()
data
```

```python
import re
p = re.compile(r"<td.*?>(.*?)</td>", re.M | re.S)
```

```python
items = p.findall(data)

exchange_rates = [float(rate.strip().replace(",", "")) for i, rate in enumerate(items) if i < 3]
exchange_rates
```
